```python
import numpy as np
import pandas as pd
import pylab as plt
from matplotlib.patches import Polygon
data = pd.read_csv('dataCR.csv',sep=' ')
gal = data[data['Name']=='NGC 224']
gal_np = gal.to_numpy()

def plot_crs(galaxy):
    fig = plt.figure(figsize=[20,5])
    ax = plt.gca()
    ax.set_xlim((min(galaxy[:,2]) - 10., max(galaxy[:,2]) + 10.)) 
    ax.set_ylim((0,1))
    ax.set_title('{}'.format(galaxy[0,1]), fontsize=30)
    ax.set_xlabel('R, arcsec', fontsize=25)
    for i in range(0, len(galaxy)):
        x1 = galaxy[i,2]
        x2 = galaxy[i,2] + galaxy[i,5]
        x3 = galaxy[i,2] - galaxy[i,4]
        plt.axvline(x1)
        if (x2 != 0.):
            c = list(np.random.choice(np.arange(0.0, 1.0, 0.06), size=3))
            ax.add_patch(Polygon([[x1,0],[x2,0],[x1,1]], closed=True,fill=True, color=c, alpha = 0.5))
            ax.add_patch(Polygon([[x1,0],[x1,1],[x3,0]], closed=True,fill=True, color=c, alpha = 0.5))
        ax.scatter(x1, 0, s = 200)
    a = galaxy[:,2]
    b = np.array([min(galaxy[:,2]) - 10., max(galaxy[:,2]) + 10.])
    frame = np.hstack([a, b])
    plt.xticks(frame.tolist(), rotation = 70)
    plt.show()
```

```python
all_gal = data['Name']
all_np = all_gal.to_numpy()
done = np.empty(90, dtype=object)
for i in range(89):
    gal = all_np[i]
    if gal in done: 
        continue
    done[i] = gal
    gg = data[data['Name']== gal]
    gal_np = gg.to_numpy()
    plot_crs(gal_np)
```
